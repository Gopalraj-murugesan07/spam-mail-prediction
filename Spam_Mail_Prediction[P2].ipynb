# Spam Mail Prediction

Ex Span mail : Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005.Text FA to 87121 to receive entry question . iee....Discounts,offers.

Ham mail[Non-Spam mail] : Pls go ahead with watts.I just wanted to be sure.Do have a great weekend.Abiola i.e Personal,Professional.

WORK FLOW

Mail data -----> Data pre processing -----> Train Test split -----> Logistic regression model ----->(new mail) -> Trained Logistic regression model -----> Spam or Ham Prediction.

Importing the Dependencies :

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Pre-processing

```python
# loading the data from csv file to a pandas dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
# checking null values
raw_mail_data.isnull().sum()
```

```python
# if it contains null values
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printing first 5 rows
mail_data.head()
```

```python
# checking number of rows and columns in the dataframe
mail_data.shape
```

Label Encoding

```python
# label spam mail as 0 ; ham mail as 1 ;
mail_data.loc[mail_data['Category'] == 'spam','Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham','Category',] = 1
```

0 ---> Spam

1 ---> ham

```python
mail_data.head()
```

```python
# seperating the data as texts and label

X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to logistic regression model

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y train and Y test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

Training the Model

Logistic Regresssion

```python
model = LogisticRegression()
```

```python
# training the logistic Regression model with the training data
model.fit(X_train_features,Y_train)
```

Evaluating the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(prediction_on_training_data,Y_train)
```

```python
print("Accuracy on training Data : ",accuracy_on_training_data)
```

```python
# prediction on testing data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(prediction_on_testing_data,Y_test)
```

```python
print("Accuracy on test Data : ",accuracy_on_test_data)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction
prediction = model.predict(input_data_features)

print(prediction)

if(prediction[0] == 1):
  print('Ham mail')
else:
  print('Spam mail')
```
